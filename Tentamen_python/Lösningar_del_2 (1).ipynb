```python
# Lösning uppgift 1a
# ------------------
def pythagorean_triplets(n):
    """
    Hittar alla Pythagoreiska tripplar där a, b och c är heltal mellan 1 och n.
    En Pythagoreisk trippel uppfyller relationen c^2 = a^2 + b^2.
    Resultatet sparas i en lista och skrivs sedan ut.
    """
    triplets = []  # Lista för att spara alla tripplar
    for a in range(1, n+1):
        for b in range(a, n+1):  # Börja från 'a' för att undvika duplicering
            for c in range(b, n+1):  # Börja från 'b' för att c ska vara det största talet
                if a**2 + b**2 == c**2:  # Kontrollera Pythagoras sats
                    triplets.append([a, b, c])  # Lägg till trippeln i listan om villkoret är uppfyllt
    return triplets

#while loop för användar input
while True:
    try:
        # Ber användaren att ange värdet på N
        N = int(input("Ange talet N: "))
        if N > 0:
            break
        else:
            print("Du måste ange ett heltal större än 0.")
    except ValueError:
        print("Du måste ange ett heltal.")

# Hitta Pythagoreiska tripplar
result = pythagorean_triplets(N)

#skriver ut användarens val av tal 
print("Ange talet N: " + str(N))

# Skriv ut resultatet
print(f"Det finns {len(result)} heltalstalskombinationer som uppfyller Pythagoras sats:")
for i, triplet in enumerate(result, 1):
    print(f"{i}. {triplet}")
```

```python
# Skapar en funktion för att omvandla hexadecimala tal till binära tal
def omvandla(tal):
    hex_to_bin = {
        '0': '0000', '1': '0001', '2': '0010', '3': '0011',
        '4': '0100', '5': '0101', '6': '0110', '7': '0111',
        '8': '1000', '9': '1001', 'A': '1010', 'B': '1011',
        'C': '1100', 'D': '1101', 'E': '1110', 'F': '1111'
    }
    bin_tal = ''
    for digit in tal.upper():
        if digit in hex_to_bin:
            bin_tal += hex_to_bin[digit]
        else:
            bin_tal += '????'  # För ogiltiga tecken
    return bin_tal

# while loop för användar input
while True:
    hex_tal = input("Ange ett hexadecimalt tal som ska omvandlas: ").upper()
    ogiltiga_tecken = [char for char in hex_tal if char not in "0123456789ABCDEF"]
    if len(ogiltiga_tecken) == 0:
        break
    else:
        print("Felaktig inmatning. Ange ett giltigt hexadecimalt tal.")

# Huvudprogram som anropar funktionen omvandla() och presenterar resultatet
bin_tal = omvandla(hex_tal)
print("Ange ett hexadecimalt tal som ska omvandlas: " + hex_tal)
print("Det binära talet blir " + bin_tal)
```

```python
# Lösning uppgift 2a
# ------------------

#importar de som krävs för uppgiften
import random
import matplotlib.pyplot as plt

# Funktion för att utföra testet och rita diagram
def test_slumptal(min_val, max_val, N):
    A_vinn, B_vinn, C_vinn = 0, 0, 0
    A_vinn_andel = []

    for _ in range(N):
        # Generera fyra slumptal A, B, C och D
        A = random.uniform(min_val, max_val)
        B = random.uniform(min_val, max_val)
        C = random.uniform(min_val, max_val)
        D = random.uniform(min_val, max_val)
        
        # Hittar vilket tal som är närmast D och inkrementera motsvarande _vinn variabel
        closest = min([A, B, C], key=lambda x: abs(x - D))
        if closest == A:
            A_vinn += 1
        elif closest == B:
            B_vinn += 1
        else:
            C_vinn += 1

        # Sparar andelen A_vinn för varje steg för att kunna rita upp grafen
        A_vinn_andel.append(A_vinn / (1 + A_vinn + B_vinn + C_vinn))
    
    # Skriver ut A_vinn, B_vinn och C_vinn
    print(f"A_vinn = {A_vinn}, B_vinn = {B_vinn}, C_vinn = {C_vinn}")

    # Ritar upp diagram
    plt.plot(A_vinn_andel)
    plt.xlabel('antalet simuleringar')
    plt.ylabel('andelen A_vinn som funktion av antalet simuleringar')
    plt.grid()
    plt.show()

# Kod för att ta emot och kontrollera användarinmatning för min_val
while True:
    try:
        min_val = int(input("Ange min (mellan 1 och 100): "))
        if 1 <= min_val <= 100:
            break
        else:
            print("Felaktig inmatning. Ange ett tal mellan 1 och 100.")
    except ValueError:
        print("Du måste ange ett heltal.")

while True:
    try:
        max_val = int(input("Ange max (mellan 1 och 100): "))
        if 1 <= max_val <= 100:
            break
        else:
            print("Felaktig inmatning. Ange ett tal mellan 1 och 100.")
    except ValueError:
        print("Du måste ange ett heltal.")

while True:
    try:
        N = int(input("Ange antalet simuleringar N (mellan 1 och 1000): "))
        if 1 <= max_val <= 1000:
            break
        else:
            print("Felaktig inmatning. Ange ett tal mellan 1 och 1000.")
    except ValueError:
        print("Du måste ange ett heltal.")

# Skriv ut detaljer före simuleringens resultat
print("Ange min: " + str(min_val))
print("Ange max: " + str(max_val))
print("Ange antalet simuleringar N: " + str(N))
print("Resultat:")
print("----------")
# Använd funktionen med givna värden
test_slumptal(min_val, max_val, N)
```

```python
# Lösning uppgift 2b
# ------------------
def sort_funktion(in_lista):                                            
    n=len(in_lista)
    for i in range(n):
        # Ändrade range(0, n) till range(1, n) för att undvika indexfel.
        for j in range(1, n): # Ändrat för att inte gå utanför index
            # Ändrade j+1 till j och j till j-1 för att jämföra intilliggande element
            # och korrigera index för korrekt jämförelse.
            if in_lista[j-1] > in_lista[j]: # Ändrat för korrekt jämförelse av element
                # Ändrade == till = för att korrekt tilldela värden.
                in_lista[j-1], in_lista[j] = in_lista[j], in_lista[j-1] # Ändrat till korrekt tilldelning
    return in_lista # Rättat namnet på returvariabeln

test_lista= [6, 37, 25, 12, 4, 1, 9]
sort_funktion(test_lista)
```

```python
# Lösning uppgift 3a
# ------------------

#importar csv för att kunna läsa csv
import csv

# Definiera en funktion för att läsa in en csv-fil
def read_file(filename):
    with open(filename, 'r', encoding='utf-8') as file:
        reader = csv.reader(file, delimiter=';')
        data = []
        for line in reader:
            data.append(line)
    return data
# anropar funktionen för filen och sparar data
kpiData = read_file("kpi.csv")

for row in kpiData[:3]:
    print(row)
```

```python
# Lösning uppgift 3b
# ------------------


# Definerar funktion för minsta värdet
def min_value(numbers):
    # Initialisera minimum till det första numret i listan
    minimum = numbers[0]
    for number in numbers:
        if number != "" and number < minimum:
            minimum = number
    return minimum

# Definerar funktion för största värdet
def max_value(numbers):
    # Initialisera maximum till det första numret i listan
    maximum = numbers[0]
    for number in numbers:
        if number != "" and number > maximum:
            maximum = number
    return maximum

# Definerar funktion för medelvärdet
def medel(numbers):
    total = 0
    count = 0
    for number in numbers:
        # Försöker att konvertera varje nummer till en float om det inte är None
        if number is not None:
            try:
                number = float(number)
                total += number
                count += 1
            except ValueError:
                # Om konvertering misslyckas, hoppar över detta nummer
                continue
    return total / count if count != 0 else 0


# Funktion för att beräkna årlig inflationstakt
def beräkna_årlig_inflation(kpiData):
    # Skapar en tom lista för att lagra årliga inflationstakter
    inflation_rates = []
    # Loopar igenom åren i KPI-data, börjar från det tredje elementet (andra året med data)
    for i in range(2, len(kpiData)):
        # Extraherar KPI-värden för det föregående året och det aktuella året, exkluderar None-värden
        previous_year_kpis = [value for value in kpiData[i-1][1:] if value is not None]
        current_year_kpis = [value for value in kpiData[i][1:] if value is not None]

        # Hoppar över beräkningen om det inte finns tillräckligt med datapunkter
        if not previous_year_kpis or not current_year_kpis:
            continue

        # Beräknar genomsnittlig KPI för det föregående året och det aktuella året
        avg_previous_year_kpi = medel(previous_year_kpis)
        avg_current_year_kpi = medel(current_year_kpis)

        # Beräknar inflationstakten och lägger till den i listan
        inflation_rate = ((avg_previous_year_kpi - avg_current_year_kpi) / avg_current_year_kpi) * 100

        inflation_rates.append((kpiData[i-1][0], inflation_rate))

    # Skapa separata listor med bara inflationstakter för jämförelse
    årliga_inflationstakter = [rate[1] for rate in inflation_rates]

    # Hittar index för året med högsta och lägsta inflationstakt
    högsta_inflation_index = årliga_inflationstakter.index(max_value(årliga_inflationstakter))
    lägsta_inflation_index = årliga_inflationstakter.index(min_value(årliga_inflationstakter))

    # Hämtar hela tuplerna för högsta och lägsta inflationstakt
    högsta_inflation = inflation_rates[högsta_inflation_index]
    lägsta_inflation = inflation_rates[lägsta_inflation_index]


    return högsta_inflation, lägsta_inflation, inflation_rates



# Funktion för att beräkna månatlig inflationstakt
def beräkna_månadlig_inflation(kpiData):
    # Skapar en tom lista för att lagra månatliga inflationstakter
    månatliga_inflationstakter = []
    
    # Loopar igenom åren i KPI-data
    for i in range(1, len(kpiData)):
        for j in range(1, 12):
            # Kontrollerar om den nuvarande månaden och nästa månad har KPI-värden
            if kpiData[i][j] is not None and kpiData[i][j+1] is not None:
                # Beräknar inflationstakten mellan de två månaderna
                nuvarande_månad_kpi = kpiData[i][j]
                nästa_månad_kpi = kpiData[i][j+1]
                inflationstakt = ((nästa_månad_kpi - nuvarande_månad_kpi) / nuvarande_månad_kpi) * 100
                månatliga_inflationstakter.append((f"{kpiData[i][0]}/{kpiData[0][j]}", inflationstakt))

    # Skapa separata listor med bara inflationstakter för jämförelse
    månatliga_inflationstakter_värden = [rate[1] for rate in månatliga_inflationstakter]

    # Hittar index för månaden med högsta och lägsta inflationstakt
    högsta_inflation_index = månatliga_inflationstakter_värden.index(max_value(månatliga_inflationstakter_värden))
    lägsta_inflation_index = månatliga_inflationstakter_värden.index(min_value(månatliga_inflationstakter_värden))

    # Hämtar hela tuplerna för högsta och lägsta inflationstakt
    högsta_inflation = månatliga_inflationstakter[högsta_inflation_index]
    lägsta_inflation = månatliga_inflationstakter[lägsta_inflation_index]


    return högsta_inflation, lägsta_inflation

# Funktion för att beräkna sammansatt genomsnittlig årlig inflationstakt
def beräkna_sammansatt_genomsnittlig_årlig_inflation(kpiData):
    # Rensar och konverterar KPI-värden till float
    kpi_2023 = []
    for kpi in kpiData[1][1:]:
        try:
            # Försök konvertera till float och lägg till i listan om det inte är None
            if kpi is not None and kpi.strip() != "":
                kpi_2023.append(float(kpi))
        except ValueError:
            # Om konverteringen misslyckas, fortsätt med nästa värde
            continue

    # Beräkna genomsnittlig KPI för 2023 med tillgängliga data
    genomsnittlig_kpi_2023 = medel(kpi_2023)

    # Använd genomsnittlig KPI för 1980 och 2023 för att beräkna den sammansatta genomsnittliga årliga inflationstakten
    genomsnittlig_kpi_första_året = medel([float(kpi) for kpi in kpiData[-1][1:] if kpi is not None])
    total_kpi_förändring = (genomsnittlig_kpi_2023 - genomsnittlig_kpi_första_året) / genomsnittlig_kpi_första_året
    totala_år = 2023 - 1980
    genomsnittlig_årlig_inflationstakt = ((1 + total_kpi_förändring) ** (1 / totala_år)) - 1
    return genomsnittlig_årlig_inflationstakt * 100  # Konvertera till procent

# Beräkna den genomsnittliga årliga inflationstakten över hela perioden
genomsnittlig_årlig_inflationstakt = beräkna_sammansatt_genomsnittlig_årlig_inflation(kpiData)

# Formatera den genomsnittliga årliga inflationstakten med en decimal
formaterad_genomsnittlig_årlig_inflationstakt = round(genomsnittlig_årlig_inflationstakt, 1)

# Läser in KPI-data från CSV-filen
kpiData = read_file("kpi.csv")

# Justera koden för att säkerställa korrekt konvertering från sträng till float och hantering av tomma strängar
for i in range(1, len(kpiData)):
    for j in range(1, len(kpiData[i])):
        try:
            kpiData[i][j] = float(kpiData[i][j]) if kpiData[i][j] else None
        except ValueError:
            # Om konvertering till float misslyckas, sätt värdet till None
            kpiData[i][j] = None



# Utskriftsfunktion 
def print_inflation_report(highest_annual, lowest_annual, highest_monthly, lowest_monthly, average_inflation):
    header = "Inflationen i Sverige under tidsperioden 1 januari 1980 -- september 2023 uttryckt i %"
    underline = "-" * 120
    
    titles = "{:<30}{:<15}{:<15}{:<15}{:<15}{:>20}".format(
        "", "Årlig", "", "Månad", "", "Medelinflation under perioden (%)"
    )
    
    sub_titles = "{:<25}{:<15}{:<15}{:<15}{:<15}".format(
        "", "år Högst", "år Lägst", "år/mån Högst", "år/mån Lägst"
    )
    
    annual_data = "{:<25}{:<14} {:<6}{:<6}{:<6}{:>50}".format(
        "Inflationstakt årlig:", 
        "{} {:.1f}".format(highest_annual[0], highest_annual[1]), 
        "{} {:.1f}".format(lowest_annual[0], lowest_annual[1]), 
        "", 
        "", 
        "{:.1f}".format(average_inflation)
    )
    
    monthly_data = "{:<55}{:<15}{:<15}".format(
        "Inflationstakt månatlig:", 
        "{} {:.1f}".format(highest_monthly[0], highest_monthly[1]), 
        "{} {:.1f}".format(lowest_monthly[0], lowest_monthly[1])
    )

    # Skriv ut rubrik och titlar
    print(header)
    print(underline)
    print(titles)
    print(sub_titles)
    print(underline)
    
    # Skriv ut årlig inflationstaktsdata
    print(annual_data)

    # Skriv ut månatlig inflationstaktsdata
    print(monthly_data)

# Kör nu beräkningar av inflationstakten igen
högsta_årlig_inflation, lägsta_årlig_inflation, inflation_rates = beräkna_årlig_inflation(kpiData)
högsta_månadlig_inflation, lägsta_månadlig_inflation = beräkna_månadlig_inflation(kpiData)

# Formatera den högsta och lägsta årliga inflationstakten med en decimal
formaterad_högsta_årlig_inflation = (högsta_årlig_inflation[0], round(högsta_årlig_inflation[1], 1))
formaterad_lägsta_årlig_inflation = (lägsta_årlig_inflation[0], round(lägsta_årlig_inflation[1], 1))

# Formatera den högsta och lägsta månatliga inflationstakten med en decimal
formaterad_högsta_månadlig_inflation = (högsta_månadlig_inflation[0], round(högsta_månadlig_inflation[1], 1))
formaterad_lägsta_månadlig_inflation = (lägsta_månadlig_inflation[0], round(lägsta_månadlig_inflation[1], 1))

# Kör utskriftsfunktionen med de formaterade värdena
print_inflation_report(formaterad_högsta_årlig_inflation, formaterad_lägsta_årlig_inflation, formaterad_högsta_månadlig_inflation, formaterad_lägsta_månadlig_inflation, formaterad_genomsnittlig_årlig_inflationstakt)
```
